```python
import pandas as pd
import seaborn as sns
```

```python
df_train = pd.read_csv('../dataset/train_cleaned_outliers_imputed.csv')
```

```python
numerical_columns = df_train.select_dtypes(include="float64").columns.tolist()
```

```python
df_temp = df_train[numerical_columns + ["CO2 Emissions(g/km)"]]
sns.set(style="ticks", color_codes=True)

sns.pairplot(df_temp, hue="CO2 Emissions(g/km)", vars=numerical_columns)
```
