```python
import pandas as pd
```

```python
df_train = pd.read_csv('./dataset/train_fix.csv')
df_test = pd.read_csv('./dataset/test_fix.csv')
```

```python
df_train.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for column in df_train.columns:
    if df_train[column].dtype == "object":
        print(column)
        sns.countplot(x=column, data=df_train)
        plt.show()
```

```python
for column in df_test.columns:
    if df_test[column].dtype == "object":
        print(column)
        sns.countplot(x=column, data=df_test)
        plt.show()
```
