# Preprocessing

Numerical Value:
- Fuel Consumption City (L/100Km) -> log, sqrt, boxcox, inverse
- Fuel Consumption Hwy (L/100Km) -> log, sqrt, boxcox, inverse
- Fuel Consumption Comb (L/100Km) -> log, sqrt, boxcox, inverse
- Engine Size(L) -> log, sqrt, boxcox, inverse
- Cylinders -> log, sqrt, boxcox, inverse

Categorical Value:
- Make -> one-hot
- Vehicle Class -> binary
- Transmission -> binary 
- Fuel Type -> one-hot

```python
from sklearn.preprocessing import OneHotEncoder
from category_encoders import BinaryEncoder


def encode_categorical_features(
    df_train, df_test, categorical_features_onehot, categorical_features_binary
):
    df_train_categorical_one_hot = df_train[categorical_features_onehot]
    df_train_categorical_binary = df_train[categorical_features_binary]

    df_test_categorical_one_hot = df_test[categorical_features_onehot]
    df_test_categorical_binary = df_test[categorical_features_binary]

    encoder_onehot = OneHotEncoder(sparse_output=False)
    encoder_onehot.set_output(transform="pandas")
    train_onehot_encoded_data = encoder_onehot.fit_transform(
        df_train_categorical_one_hot
    )
    test_onehot_encoded_data = encoder_onehot.transform(df_test_categorical_one_hot)

    encoder_binary = BinaryEncoder(cols=categorical_features_binary)
    encoder_binary.set_output(transform="pandas")
    train_df_binary = encoder_binary.fit_transform(df_train_categorical_binary)
    test_df_binary = encoder_binary.transform(df_test_categorical_binary)

    onehot_encoded_data_cat_train = (
        train_onehot_encoded_data.astype("int64")
    )
    df_binary_cat_train = train_df_binary.astype("int64")

    onehot_encoded_data_cat_test = (
        test_onehot_encoded_data.astype("int64")
    )
    df_binary_cat_test = test_df_binary.astype("int64")

    # Merge the one-hot and binary encoded dataframes with the original dataframes
    df_train = pd.concat(
        [
            df_train.drop(
                categorical_features_onehot + categorical_features_binary, axis=1
            ),
            onehot_encoded_data_cat_train,
            df_binary_cat_train,
        ],
        axis=1,
    )
    df_test = pd.concat(
        [
            df_test.drop(
                categorical_features_onehot + categorical_features_binary, axis=1
            ),
            onehot_encoded_data_cat_test,
            df_binary_cat_test,
        ],
        axis=1,
    )

    return df_train, df_test


onehot_columns = ["Make", "Fuel Type"]
binary_columns = ["Vehicle Class", "Transmission"]
```

```python
from sklearn.preprocessing import PowerTransformer, QuantileTransformer # <- might use this too


def power_transform_dataframe(df_train, df_test, columns_to_transform):
    pt = PowerTransformer(standardize=True)
    transformed_data_train = pt.fit_transform(df_train[columns_to_transform])
    transformed_df_train = pd.DataFrame(
        transformed_data_train, columns=columns_to_transform
    )
    df_train.loc[:, columns_to_transform] = transformed_df_train

    transformed_data_test = pt.transform(df_test[columns_to_transform])
    transformed_df_test = pd.DataFrame(
        transformed_data_test, columns=columns_to_transform
    )
    df_test.loc[:, columns_to_transform] = transformed_df_test

    return df_train, df_test


columns_to_power_transform = [
    "Fuel Consumption City (L/100Km)",
    "Fuel Consumption Hwy (L/100Km)",
    "Fuel Consumption Comb (L/100Km)",
    "Engine Size(L)",
    "Cylinders",
]
```

```python
from sklearn.model_selection import KFold, cross_val_score

random_state = 42
kf = KFold(n_splits=5, shuffle=True, random_state=random_state)
y = df_target

for train_index, test_index in kf.split(df_train_fix):
    X_train, X_test = df_train_fix.iloc[train_index], df_train_fix.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Normalize the data
    X_train, X_test = power_transform_dataframe(
        X_train, X_test, columns_to_power_transform
    )
    # X_train, X_test = encode_categorical_features(
    #     X_train, X_test, onehot_columns, binary_columns
    # )

    print(X_train.isna().sum())
    print(X_test.isna().sum())
    
    print(X_train)
```
